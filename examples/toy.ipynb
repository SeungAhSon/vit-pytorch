```python
a = torch.rand(5, 2, 3, 4)
print(a.shape)
print(a)
```

```python
import torch

# Get the size of the last two dimensions
rows, cols = a.shape[-2:]

# Create an index tensor for the diagonal elements
diag_idx = torch.arange(min(rows, cols))

# Set the diagonal elements to zero
a[..., diag_idx, diag_idx] = 0

print(a)
```

```python
print(a.fill_diagonal_(0))
```

```python
import torch

# Define the size of the square matrix
matrix_size = 5

# Create an identity matrix
identity_matrix = torch.eye(matrix_size)
print(identity_matrix)

# Create a tensor with zero diagonal elements
zero_diagonal_matrix = identity_matrix - torch.diag(torch.ones(matrix_size))
print(torch.diag(torch.ones(matrix_size)))

print(zero_diagonal_matrix)
```

```python
import torch

# Define the size of the square matrix
matrix_size = 5

# Create a tensor with non-zero values
non_zero_matrix = torch.rand((matrix_size, matrix_size+2))
print(non_zero_matrix)

# Set the diagonal elements to zero
non_zero_matrix.fill_diagonal_(0)

print(non_zero_matrix)
```

```python
import pickle

# 파일 불러오기
with open("RLPE.pkl", "rb") as f:
    RLPE = pickle.load(f)
```

```python
print(RLPE.shape)
print(RLPE)

print(RLPE.max())
print(RLPE.min())

def min_max(x):
  big = x.max()
  small = x.min()
  
  new_x = (x - small) / (big - small)
  new_x = (new_x - 0.5) * 2
  return new_x

scaling_RLPE = min_max(RLPE)
print(scaling_RLPE.max())
print(scaling_RLPE.min())
```

```python
# 파일로 저장
with open("scaling_RLPE.pkl", "wb") as f:
    pickle.dump(scaling_RLPE, f)
```
